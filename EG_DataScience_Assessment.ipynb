# Exploratory Data Analysis

One thing I noticed when I first loaded in the dataset is that there were missing values in the Age, HoursPerWeek, and TotalHours columns.

```python
for column in list(sc_data.drop(['LeagueIndex', 'GameID'], axis=1).columns):
    for i in range(1,9):
        temp = sc_data[sc_data['LeagueIndex'] == i]
        if '?' in temp[column].unique():
            print("There is a ? in " + column + " at LeagueIndex " + str(i))
            print(temp[column].value_counts())
```

```python
y = sc_data[sc_data['LeagueIndex'] == 5]
y[y['TotalHours'] == '?']
```

```python
dia = sc_data[sc_data['LeagueIndex'] == 5]
dia = dia.drop(1841)
sc_data.loc[1841, ['HoursPerWeek']] = [dia['HoursPerWeek'].median()]
dia = dia.drop(358)
sc_data.loc[1841, ['TotalHours']] = [dia['TotalHours'].median()]
```

```python
gm = sc_data[sc_data['LeagueIndex'] == 7]
sc_data['Age'].replace(to_replace="?", value=gm['Age'].median(), inplace = True)
sc_data['HoursPerWeek'].replace(to_replace="?", value=gm['HoursPerWeek'].median(), inplace = True)
sc_data['TotalHours'].replace(to_replace="?", value=gm['TotalHours'].median(), inplace = True)
sc_data['Age'] = pd.to_numeric(sc_data['Age'])
sc_data['HoursPerWeek'] = pd.to_numeric(sc_data['HoursPerWeek'])
sc_data['TotalHours'] = pd.to_numeric(sc_data['TotalHours'])
sc_data = sc_data.drop('GameID', axis=1)
sc_data[sc_data['LeagueIndex'] == 8]
```

After further investigation, these values were missing from all players whose LeagueIndex was 8 (Professional), as well as two Diamond players. There are two ways to solve this problem: removing the Age, HoursPerWeek, and TotalHours variables from the dataset, or replacing the question marks with some sort of center measure (mean, median, or mode). Creating a model to predict the missing values of Age, HoursPerWeek, and TotalHours is a good idea but cannot be applied in this case, since there would be no training data for those whose LeagueIndex = 8.

Thus, for now, I will have two datasets: One that removes Age, HoursPerWeek, and TotalHours, and one that fills the missing values, and see which one has the better performance. For the filling of the missing values, I filled the Diamond players with the median HoursPerWeek and Age for LeagueIndex = 5, and for the Professional players, I filled with the median HoursPerWeek, Age, and TotalHoursPlayed for LeagueIndex = 7 (since it is the closest skill level and is a more accurate representation than the median of the entire dataset). Additionally, since HoursPerWeek, Age, and TotalHoursPlayed were "object" data types (because of the presence of the '?'s), I changed them to float64s so that they can be viewed in the correlation matrix.

# Preprocessing

```python
X = sc_data.drop('LeagueIndex', axis=1)
y = sc_data['LeagueIndex']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_preprocessed = pd.DataFrame(X_scaled, columns=X.columns)
print(X_preprocessed.head())
```

I used sklearn's StandardScaler to make all the predictors have mean 0 and variance 1 (Standard Normal Distribution). This will aid in the modeling stage.

# Modeling

## Linear Regression

```python
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed.drop(['Age', 'HoursPerWeek', 'TotalHours'], axis=1), y, test_size=0.2, random_state=42)

train_data = lgb.Dataset(X_train, label=y_train)

param_grid = {
    'max_depth': [3, 4, 5],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [100, 200, 300],
}


lgb_regressor = lgb.LGBMRegressor()


grid_search = GridSearchCV(estimator=lgb_regressor, param_grid=param_grid, scoring='neg_mean_squared_error', cv=3)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)

mse = mean_squared_error(y_test, y_pred.round())
r2 = r2_score(y_test, y_pred.round())

print("Best Parameters:", best_params)
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)

train_data = lgb.Dataset(X_train, label=y_train)

param_grid = {
    'max_depth': [3, 4, 5],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [100, 200, 300],
}


lgb_regressor = lgb.LGBMRegressor()


grid_search = GridSearchCV(estimator=lgb_regressor, param_grid=param_grid, scoring='neg_mean_squared_error', cv=3)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)

mse = mean_squared_error(y_test, y_pred.round())
r2 = r2_score(y_test, y_pred.round())

print("Best Parameters:", best_params)
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

I began the modeling process with a regression model (from the LightGBM library) as a baseline. Since regression models output a continuous value, and LeagueIndex is ordinal, I rounded the output from the model to the nearest integer. I also conducted a GridSearch to tune the hyperparameters of the LGBMRegressor.

The first model runs on the dataset with Age, HoursPerWeek, and TotalHours removed, and the second model includes all the predictors. The second model had a higher R^2 of 0.5813122302269526, meaning that ~58% of the variation in LeagueIndex is explained by all 18 predictors. With this information, for the following models, I concluded that using the dataset with all predictors will be the better option.

## sklearn Classifiers

```python
classifiers = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    SVC(),
    KNeighborsClassifier(),
    MLPClassifier(),
    GaussianNB()
]


for classifier in classifiers:
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    
    print(type(classifier).__name__)
    print("Accuracy:", accuracy)
    print("Classification Report:")
    print(report)
    print("-----------------------------------------")
```

Next, I fit several classification models from the sklearn library. I fit seven in total:  Logistic Regression, DecisionTree, RandomForest, SVC (Support Vector Classifier), KNeighbors, MLP (Multi-layer Perceptron), and GaussianNB (Naive Bayes). Of the seven, the MLP model had the highest test accuracy of 0.42.

## Other Classifiers: LightGBM and XGBoost

```python
#NOTE: In order to run the LightGBM and XGBoost classifiers, labels must be reassigned from 1-8 to 0-7 
#(the classification remains the same, since the mapping is one-to-one)
temp = y-1

X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, temp, test_size=0.2, random_state=42)

dtrain = xgb.DMatrix(X_train, label=y_train)

param_grid = {
    'max_depth': [3, 4, 5],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [100, 200, 300],
}


xgb_classifier = xgb.XGBClassifier(objective='multi:softmax', num_class=len(y.unique()))


grid_search = GridSearchCV(estimator=xgb_classifier, param_grid=param_grid, scoring='accuracy', cv=3)
grid_search.fit(X_train, y_train)


best_params = grid_search.best_params_
best_model = grid_search.best_estimator_


y_pred = best_model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("Best Parameters:", best_params)
print("Accuracy:", accuracy)
print("Classification Report:")
print(report)
```

```python
train_data = lgb.Dataset(X_train, label=y_train)

params = {
    'objective': 'multiclass',
    'num_class': len(y.unique()),
    'metric': 'multi_logloss',
    'num_leaves': 20
}


model = lgb.train(params, train_data)

y_pred = model.predict(X_test)
y_pred_class = y_pred.argmax(axis=1)

accuracy = accuracy_score(y_test, y_pred_class)
report = classification_report(y_test, y_pred_class)

print("Accuracy:", accuracy)
print("Classification Report:")
print(report)
```

With 0.42 being my highest test accuracy so far, I looked to improve this score with models outside of sklearn. Thus, I explored classifiers from LightGBM and XGBoost, both of which are frameworks for gradient-boosted trees. I performed a GridSearch on both the LightGBM and XGBoost classifiers and yet again my test accuracy peaked at 0.42 with the LightGBM classifier.

After building several models, the model I am most confident in is th LightGBM classifier with an accuracy of 0.42. Even though the MLP model had the same accuracy, the LightGBM model had higher precision, recall, and f1-scores.
